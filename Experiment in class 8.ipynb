# Plotting charts without using dataframes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Storing values to be plotted in a list
Income = [1350000, 2150000, 1570000, 1730000, 2010000, 3210000, 5540000]
Expenditure = [950000, 1850000, 1350000, 1400000, 2900000, 3650000, 1730000]
Year = [2005, 2006, 2007, 2008, 2009, 2010, 2011]

# Calculating Surplus from the difference between Income and Expenditure
Surplus = []
for x in range(len(Income)):
    a = Income[x] - Expenditure[x]
    Surplus.append(a)
```

```python
plt.rcParams["figure.figsize"]=(7,7)
plt.rcParams["font.size"] = 16
```

```python
# plt.style.use("fivethirtyeight")
#  Setting plot figure size
plt.figure(figsize=(7,7))

# Plotting each lines, passing the year as x-axis and the other list [Income, Expenditure and Surplus] as Y-axis
# I passed the colors, linestyle, marker and label also

plt.plot(Year, Income, color = "green", label="Income")
plt.plot(Year, Expenditure, color = "red", label="Expenditure")
plt.plot(Year, Surplus, color="k", linestyle = "--", marker= "x", label="Surplus")

# Giving plot a title passing some font properties also as arguments to beautify the text
plt.title("Company XYZ yearly cashflow", fontsize=18, fontweight="bold", fontfamily = "bayou cowboy", color="b")
plt.xlabel("Year")
plt.ylabel("Income | Expenditure | Surplus")

# Legend is put on
plt.legend()

# Minor ticks is put on
plt.minorticks_on()

# Grid is put on for both axis ticks (minor & major)
plt.grid(which="both")

# for solving padding issues
plt.tight_layout()
```

## Plotting trigonometry graph

```python
# documentation for linspace
help(np.linspace)
```

```python
# Setting values for the plot

x = np.linspace(0, np.pi*5, 100)
y = np.sin(x)
```

```python
plt.plot(x, y, marker="x")
plt.fill_between(x, y, where=(y>0), interpolate = True, alpha=0.3)
plt.fill_between(x, y, where=(y<=0), interpolate = True, alpha=0.3, color="k", label="")

```

```python
plt.plot(x, y, marker="x")
plt.fill_between(x, y, alpha=0.5, where=(y>0), color="g", label="Y is greater than 0")
plt.fill_between(x, y, alpha=0.3, where=(y<0), color="r", label="Y is greater than 0")

plt.legend(loc=(1,0))
plt.tight_layout()
```
