```python
import matplotlib.pyplot as plt 
import pandas as pd
#from matplotlib import pyplot as plt

filex = r"C:\Users\Janet\Documents\DA_materials\ca-covid.csv"
df = pd.read_csv(filex)
df.drop('state', axis=1, inplace=True)
df['date'] = pd.to_datetime(df['date'], format="%d.%m.%y")
df["month"] = df["date"].dt.month
df.set_index('date',inplace=True)
```

```python
df[df["month"]==6]["cases"].plot(kind="box")
```

```python
df[df["month"]==6]["cases"].plot(kind="hist")
```

```python
df[df["month"]==6][["cases","deaths"]].plot(kind="area",stacked=False,color=['b','r'])
#by default the stack for area is FALSE
```

'-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

```python
df[df["month"]==6][["cases","deaths"]].plot(linestyle='dotted')
```

```python
#scatter plot must come with an X axis and Y axis
df[df["month"]==6][["cases","deaths"]].plot(kind="scatter",x="cases",y="deaths",color="b")
```

```python
#PIE CHART
df.groupby("month")["cases"].sum().plot(kind="pie")
```
